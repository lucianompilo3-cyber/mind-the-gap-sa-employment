```python
url = "https://raw.githubusercontent.com/statssa/QLFS/main/data/qlfs_2024.csv"

df = pd.read_csv(url)
print(df.head ())
print(df .shape)
```

```python
import pandas as pd

data = {
    'Province': ['Western Cape', 'Eastern Cape', 'Northern Cape',
                 'Free State', 'KwaZulu-Natal', 'North West',
                 'Gauteng', 'Mpumalanga', 'Limpopo'],
    'Unemployment_Q1_2025': [19.6, 39.3, 29.5, 37.9, 32.3, 40.4, 34.7, 35.4, 33.3],
    'Unemployment_Q4_2025': [18.1, 42.5, 27.1, 37.2, 32.3, 35.1, 33.0, 32.3, 28.2],
    'Unemployment_Q1_2026': [19.6, 44.6, 30.4, 37.8, 31.2, 35.3, 34.1, 36.3, 31.7]
}

df = pd.DataFrame(data)
print(df)
print("\nShape:", df.shape)
```

```python
# Which province has the highest unemployment in Q1 2026?
highest = df.loc[df['Unemployment_Q1_2026'].idxmax()]
lowest = df.loc[df['Unemployment_Q1_2026'].idxmin()]

print("Highest unemployment Q1 2026:")
print(highest['Province'], "-", highest['Unemployment_Q1_2026'], "%")

print("\nLowest unemployment Q1 2026:")
print(lowest['Province'], "-", lowest['Unemployment_Q1_2026'], "%")

print("\nNational average Q1 2026:")
print(round(df['Unemployment_Q1_2026'].mean(), 1), "%")
```

```python
import matplotlib.pyplot as plt

provinces = df['Province']
unemployment = df['Unemployment_Q1_2026']

plt.figure(figsize=(10, 6))
plt.barh(provinces, unemployment, color='steelblue')
plt.axvline(x=32.7, color='red', linestyle='--', label='National rate (32.7%)')
plt.xlabel('Unemployment Rate (%)')
plt.title('Unemployment Rate by Province - Q1 2026')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Add a column showing change from Q1 2025 to Q1 2026
df['Change'] = df['Unemployment_Q1_2026'] - df['Unemployment_Q1_2025']

# Who got worse and who got better?
print("Provinces where unemployment INCREASED:")
print(df[df['Change'] > 0][['Province', 'Change']].to_string(index=False))

print("\nProvinces where unemployment DECREASED:")
print(df[df['Change'] < 0][['Province', 'Change']].to_string(index=False))
```

```python
print("Provinces with NO change:")
print(df[df['Change'] == 0][['Province', 'Change']].to_string(index=False))
```

```python
# Chart showing which provinces got better or worse
colors = ['red' if x > 0 else 'green' if x < 0 else 'gray'
          for x in df['Change']]

plt.figure(figsize=(10, 6))
plt.barh(df['Province'], df['Change'], color=colors)
plt.axvline(x=0, color='black', linewidth=0.8)
plt.xlabel('Change in Unemployment Rate (percentage points)')
plt.title('Year-on-Year Change in Unemployment by Province\nQ1 2025 vs Q1 2026')
plt.tight_layout()
plt.show()
```

```python
df.to_csv('qlfs_clean_q1_2026.csv', index=False)
print("File saved!")
```

## Key findings


1. Eastern Cape has the highest unemployment in SA at 44.6% (Q1 2026)
2. Western Cape has the lowest at 19.6% - a 25% point gap
3. North West improved the most (-5.1 percentage points)
4. Eastern Cape worsened the most (-5.1 percentage points)
